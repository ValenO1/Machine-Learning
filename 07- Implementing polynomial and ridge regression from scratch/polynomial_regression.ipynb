```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
from linear_model import RidgeRegression
```

```python
data = pd.read_csv("polynomial_data.csv")
data
```

```python
X = data[["x"]]
y = data["y"]
plt.scatter(X,y)
```

```python
model = LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python
y_pred = model.predict(X)
plt.scatter(X,y)
plt.plot(X,y_pred)
```

```python
poly = PolynomialFeatures(include_bias=False, degree=10)
new_x = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(new_x, y)
model.score(new_x, y)
```

```python
X_range = np.linspace(0,2, 100).reshape(100,1)
X_range_poly = poly.fit_transform(X_range)
```

```python
y_pred = model.predict(X_range_poly)
plt.scatter(X,y)
plt.plot(X_range, y_pred, c="red")
```

```python
model = RidgeRegression(alpha=2)
model.fit(new_x, y)
model.score(new_x, y)
model.betas
```

```python
y_pred = model.predict(X_range_poly)
plt.scatter(X,y)
plt.plot(X_range, y_pred, c="red")
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=2)
model.fit(new_x, y)
model.score(new_x, y)
model.coef_
```
