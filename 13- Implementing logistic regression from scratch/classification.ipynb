```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from linear_model import LogisticRegression as my_model
```

```python
data = pd.read_csv("exams.csv")
data
```

```python
X = data[['score']]
y = data['admission']
plt.scatter(X,y)
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
model = my_model(learning_rate=.1, iterations= 1000)
model.fit(X,y)
model.score(X,y)
```

```python
X_range = np.linspace(0, 1, 100).reshape(-1,1)
plt.scatter(X,y)
plt.plot(X_range, model.predict_proba(X_range))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="sag")
model.fit(X,y)
model.score(X,y)
```

```python
from sklearn.linear_model import SGDClassifier
model = SGDClassifier(eta0=.1, random_state=42)
model.fit(X,y)
model.score(X,y)
```
