```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from svm import LinearSVM
```

```python
data = pd.read_csv("separable data.csv")
data
```

```python
X = data.drop("class", axis=1).values
y = data["class"].values
y = np.where(y == 1, 1, -1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
model = LinearSVM()
model.fit(X,y)
model.score(X,y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, cmap= plt.cm.coolwarm)
plt.xlabel("feature 1")
plt.ylabel("feature 2")
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.linspace(x_min,x_max, 1000), np.linspace(y_min,y_max, 1000))

z = model.predict(np.c_[xx.flatten(), yy.ravel()])
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, alpha = .2, cmap = plt.cm.coolwarm)
```

```python
model2 = LinearSVC(dual= False)
model2.fit(X,y)
model2.score(X,y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, cmap= plt.cm.coolwarm)
plt.xlabel("feature 1")
plt.ylabel("feature 2")
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.linspace(x_min,x_max, 1000), np.linspace(y_min,y_max, 1000))

z = model2.predict(np.c_[xx.flatten(), yy.ravel()])
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, alpha = .2, cmap = plt.cm.coolwarm)
```

```python
model3 = SGDClassifier(loss="hinge")
model3.fit(X,y)
model3.score(X,y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, cmap= plt.cm.coolwarm)
plt.xlabel("feature 1")
plt.ylabel("feature 2")
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.linspace(x_min,x_max, 1000), np.linspace(y_min,y_max, 1000))

z = model3.predict(np.c_[xx.flatten(), yy.ravel()])
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, alpha = .2, cmap = plt.cm.coolwarm)
```
